```python
from nltk.corpus import stopwords 
from nltk.stem.wordnet import WordNetLemmatizer
import string
```

```python
def clean(doc):
    stop_free = " ".join([i for i in doc.lower().split() if i not in stop])
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    return normalized
```

```python
stop = set(stopwords.words('english'))
exclude = set(string.punctuation) 
lemma = WordNetLemmatizer()
```

```python
doc_clean = [clean(doc).split() for doc in data]
```

```python
doc_clean[0]
```

```python
doc_clean[1]
```

```python
# Gensim
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel
```

```python
#To run any mathematical model on text corpus, it is a good practice to convert it into a matrix representation. LDA model looks for repeating term patterns in the entire DT matrix
```

```python
# Creating the term dictionary of our courpus, where every unique term is assigned an index.
```

```python
# Converting list of documents (corpus) into Document Term Matrix using dictionary prepared above.
```

```python
dictionary = corpora.Dictionary(doc_clean)
```

```python
doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]
```

```python
# Creating the object for LDA model using gensim library
Lda = gensim.models.ldamodel.LdaModel
```

```python
# Running and Trainign LDA model on the document term matrix.
ldamodel = Lda(doc_term_matrix, num_topics=3, id2word = dictionary, passes=50)
```

```python
print(ldamodel.print_topics(num_topics=3, num_words=15))
```
